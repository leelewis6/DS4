## Summary

Objective: Predict home prices given attributes about the homes. 

### Data Summary and Data Cleansing

- The Ames Housing Data Set has 1460 Observations and 80 predictors, exluding the target. 
- The data set was cleansed by converting categorical variables to dummy variables.
- Null values for integers were filled with the mean value.

### Exploration
- The target variable, Sales Price,was slightly skewed. I choose to remove outliers to create a distribution that's closer to normal.
- I explored GrLivArea and 1stFlrSF. Both features were highly correlated with SalePrice. 
- I removed outliers from both features. 

### Model Selection
1. Due to the large number of variables, I used Lasso regression. 
2. For comparison, I also ran an OLS regression as a baseline
3. By using CV, I found the best alpha for Lasso to be 100. 
4. Lasso yielded a slightly better RMSE. 
5. Lasso's better performance can be attributed to the removal of non-significant and collinear predictors. 
6. Lasso RMSE = 20990
7. OLS RSME = 23345

### Prepare Data for Modeling

```python
#Create Dummy Variables
df = pd.get_dummies(train)
```

```python
#Fill in null values with the mean
data = df.fillna(df.mean())
```

```python
#Prepare data for fitting
#Remove the target and the ID. The Id has no significance

X = data.drop(['Id','SalePrice'], axis = 1)
y = data.SalePrice
```

```python
#Hold out 30% of the data for testing. 

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size =.3, random_state=20)
```

### Simple linear regression

```python
# Fit linear model using OLS
lm = linear_model.LinearRegression()

result = lm.fit(X_train,y_train)
```

```python
#print 'Coefficients:', result.coef_
# print 'y-intercept:', result.intercept_
print 'R-Squared:', result.score(X_train, y_train)
```

```python
#Print out the coefficents in order of importance, descending...more important coefficient is first
coef = pd.DataFrame({'coef': result.coef_,'column': X_train.columns})
coef.sort('coef', ascending = False)
```

```python
#Predit the sales prices on the test data set
predicted = result.predict(X_test)

#create a dataframe to store the test and predicted results
predict_linear = pd.DataFrame({'test':y_test, 'predicted': predicted})

#Root Mean Squared Error for OLS - Linear Regression
print np.sqrt(sm.mean_squared_error(y_test, predicted))
```

### Lasso Regresion

```python
#Use Grid Search to find the best alpha for Lasso Regression

from sklearn import grid_search

alphas = np.logspace(-10, 10, 21)
gs = grid_search.GridSearchCV(
    estimator=linear_model.Lasso(),
    param_grid={'alpha': alphas},
    scoring='mean_squared_error')

lasso_result = gs.fit(X_train, y_train)
```

```python
#Get the best parameters for the Lasso Regression
print gs.best_estimator_
```

```python
#fit the model with the parameter from the =CV
lasso = linear_model.Lasso(alpha=100.0, copy_X=True, fit_intercept=True, max_iter=1000,
   normalize=False, positive=False, precompute=False, random_state=None,
   selection='cyclic', tol=0.0001, warm_start=False)

#Fit the model
lasso_result =lasso.fit(X_train, y_train)                              
```

```python
#print 'Coefficients:', lasso_result.coef_
#print 'y-intercept:', lasso_result.intercept_
print 'R-Squared:', lasso_result.score(X_train, y_train)
```

```python
coef = pd.DataFrame({'coef': lasso_result.coef_,'column': X_train.columns})
coefs_cols = ['column','coef']
coefs = coef[coefs_cols]
coefs.sort('coef',ascending = False)
```

```python
#Identify the features that lasso kept. 
zero = np.where(lasso_result.coef_==0)
not_zero = np.where(lasso_result.coef_!=0)

#Display most important features
indices = np.where(abs(lasso_result.coef_) >= 5000)
plt.barh(range(len(indices[0])), lasso_result.coef_[indices[0]])
plt.yticks(range(len(indices[0])), X.columns[indices[0]])
plt.title('Most imporant features')
```

```python
#Count the number of features lasso excluded 
count_zero = sum([len(t) for t in zero])
count_not_zero = sum([len(t) for t in not_zero])
print ("Features Excluded " + str(count_zero))
print ("Features To Keep " + str(count_not_zero))
```

```python
#Predict sales using lasso
lasso_predicted = lasso_result.predict(X_test)

#Create dataframe with test and predicted values
predict_sales = pd.DataFrame({'test':y_test, 'predicted': lasso_predicted})
predict_sales.head()
```

```python
#root mean squared error
np.sqrt(sm.mean_squared_error(y_test, lasso_predicted))
```
